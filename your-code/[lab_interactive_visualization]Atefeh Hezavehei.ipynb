## 1. Create an interactive bar chart showing total quantity and revenue by country (excluding United Kingdom) for the month of April 2011.

```python
data=df[df.Country!='United Kingdom']  # quitar el UK

data=data.loc[(data.InvoiceDate>='2011-4-1') & (data.InvoiceDate<='2011-4-30')]

data.head()
```

```python
data=data.groupby('Country').sum()

data=data[['Quantity', 'Revenue']]

data.iplot(kind='bar', xTitle='Country', title='Quantity and Revenue by Country')
```

## 2. Create an interactive line chart showing quantity and revenue sold to France between January 1st and May 31st 2011.

```python
data=df[df.Country=='France'] 

data=data.loc[(data.InvoiceDate>='2011-1-1') & (data.InvoiceDate<='2011-5-31')]

data.head()
```

```python
data=data.groupby("InvoiceDate").sum()

data=data[['Quantity', 'Revenue']]

data.iplot(kind='bar', xTitle='Country', title='Quantity and Revenue France')
```

## 3. Create an interactive scatter plot showing the relationship between average quantity (x-axis) and average unit price (y-axis) for the product PARTY BUNTING with the plot points color-coded by country (categories).

```python
data=df.loc[df.Description=='PARTY BUNTING']

data=data.groupby('Country').mean()

data=data[['Quantity', 'UnitPrice']]

data.reset_index(inplace=True)

data.iplot(x='Quantity', y='UnitPrice', categories='Country',
          xTitle='avg qty', yTitle='avg unit price', color='pink', title='AVG by Country')
```

## 4. Create a set of interactive histograms showing the distributions of quantity per invoice for the following countries: EIRE, Germany, France, and Netherlands.

```python
prod_lst=['JUMBO BAG RED RETROSPOT', 'CREAM HANGING HEART T-LIGHT HOLDER', 'REGENCY CAKESTAND 3 TIER']

country_lst=['EIRE', 'Germany', 'France']


data=df.loc[(df.Description.isin(prod_lst)) & (df.Country.isin(country_lst))]


data=data.pivot_table(values='Revenue',
                      columns='Description',
                      index='Country', aggfunc='sum')  # esto es igual que el groupby por pais

data.reset_index(inplace=True)

data.iplot(kind='bar', x='Country', title='Revenue by Country', 
           xTitle='Country', yTitle='Revenue')
```

## 5. Create an interactive scatter plot that plots number of invoices (x-axis) vs. number of customers (y-axis) and the plot points represent individual products. Add two sliders that control the x and y axis ranges.

```python
df.head()
```

```python
data=df.groupby('Description').count()

data=data[['InvoiceNo', 'CustomerID']]

data.reset_index(inplace=True)

data.iplot(x='InvoiceNo', y='CustomerID', categories='Description',
          xTitle='n.invocie', yTitle='n.customer', color='yellow', title='n.invoice vs n.customer')
```
