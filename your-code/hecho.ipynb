```python
import pandas as pd
import chart_studio.plotly as py
import cufflinks as cf
from ipywidgets import interact

cf.go_offline()
```

```python
df = pd.read_csv("Online Retail.csv", encoding='ISO-8859-1')
```

## 1. Create an interactive bar chart showing total quantity and revenue by country (excluding United Kingdom) for the month of April 2011.

```python
df_no_uk = df[df.Country!='United Kingdom']
df_no_uk=df_no_uk.loc[(df_no_uk.InvoiceDate>='2011-4-1') & (df_no_uk.InvoiceDate<='2011-4-30')]
df_no_uk=df_no_uk.groupby('Country').sum()
df_no_uk=df_no_uk[['Quantity', 'Revenue']]
df_no_uk.iplot(kind='bar', xTitle='Country', title='Quantity and Revenue by Country except UK')
```

## 2. Create an interactive line chart showing quantity and revenue sold to France between January 1st and May 31st 2011.

```python
France_quantity = df[df.Country == 'France']
France_quantity  = France_quantity.loc[(France_quantity.InvoiceDate>='2011-1-1') & (France_quantity.InvoiceDate<='2011-5-31')]
France_quantity =France_quantity[['Quantity']]
France_quantity ['Index'] = [i for i in range(len(France_quantity))]
France_quantity .iplot(x='Index', y='Quantity', xTitle='Record', yTitle='Quantity', color='pink', title='Quantity sold to France')
```

```python
France_revenue = df[df.Country == 'France']
France_revenue = France_revenue.loc[(France_revenue.InvoiceDate>='2011-1-1') & (France_revenue.InvoiceDate<='2011-5-31')]
France_revenue=France_revenue[['Revenue']]
France_revenue['Index'] = [i for i in range(len(France_revenue))]
France_revenue.iplot(x='Index', y='Revenue', xTitle='Record', yTitle='Revenue', color='purple', title='Revenue generated in France')
```

## 3. Create an interactive scatter plot showing the relationship between average quantity (x-axis) and average unit price (y-axis) for the product PARTY BUNTING with the plot points color-coded by country (categories).

```python
avg_quant=df.loc[df.Description=='PARTY BUNTING']

avg_quant=avg_quant.groupby('Country').mean()

avg_quant=avg_quant[['Quantity', 'UnitPrice']]

avg_quant.reset_index(inplace=True)

avg_quant.iplot(x='Quantity', y='UnitPrice', categories='Country',
          xTitle='Average quantity', yTitle='Average unit price', color='pink', title='Average by Country')
```

## 4. Create a set of interactive histograms showing the distributions of quantity per invoice for the following countries: EIRE, Germany, France, and Netherlands.

```python
products=['CREAM HANGING HEART T-LIGHT HOLDER', 'JUMBO BAG RED RETROSPOT','REGENCY CAKESTAND 3 TIER']

countries=['EIRE', 'Germany', 'France', 'Netherlands']


distribution=df.loc[(df.Description.isin(products)) & (df.Country.isin(countries))]


distribution=distribution.pivot_table(values='Quantity',
                      columns='Description',
                      index='Country', aggfunc='sum')  

distribution.reset_index(inplace=True)

distribution.iplot(kind='bar', x='Country', title='Quantity per invoice distributions', 
           xTitle='Country', yTitle='Quantity')
```

## 5. Create an interactive scatter plot that plots number of invoices (x-axis) vs. number of customers (y-axis) and the plot points represent individual products. Add two sliders that control the x and y axis ranges.

```python
df5 = df.groupby('CustomerID').count()
df5=df5[['InvoiceNo', 'Description']]
df5['index'] = [i for i in range(len(df5))]

df5.iplot(kind ='scatter', x='index', y='InvoiceNo',
          xTitle='InvoiceNo', yTitle='Customers', color='orange', title='Invoices vs Customers')
```
